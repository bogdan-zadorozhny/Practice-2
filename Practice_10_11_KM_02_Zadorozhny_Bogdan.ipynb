Декоратор повинен створювати список усіх сторінок у потрібній послідовності, причому номери сторінок кожного зошита мають зберігатись в окремому вкладеному списку. **Бонусне завдання (за бажанням)**

6. Замість функції з п. 4 створіть генератор, який буде визначати послідовність книжкових сторінок для кожного зошита. Ці рядки можна потім вивести або у циклі, або за допомогою функції next(). Функція-декоратор має працювати з цим генератором так само, як і зі звичайною функцією.

```
def checker_book(text):
    val = input(text)
    while type(val) != int:
        try:
            if int(val) > 1280 or int(val) < 1:
                val = "a"
            val = int(val)
        except:
            val = input("Enter a number of book pages (1-1280):\n")
    return int(val)


def checker_notebook(text):
    val = input(text)
    while type(val) != int:
        try:
            if int(val) != 16 and int(val) != 24 and int(val) != 32:
                val = "a"
            val = int(val)
        except:
            val = input("Enter a number of pages of each notebook(16, 24 or 32):\n")
    return int(val)


def notebooks(book, notebook):
    notebooks = int(book / notebook)
    print("Amount of notebooks in book:", notebooks)
    list_of_notebooks = []
    for j in range(notebooks):
        a = j * notebook
        a = [i for i in range(a+1, notebook+a+1)]
        b = []
        n = int(notebook / 2 - 1)
        for i in range(0, n, 2):
            b += [a[-i-1], a[i], a[i+1], a[-i-2]]
        list_of_notebooks.append(b)
    return list_of_notebooks


def decorator(choice, func):
    def additional_list(func):
        print(func)
    if choice:
        return additional_list(func)
    return print(*func, sep="\n")


while True:
    try:
        book = checker_book("Enter a number of book pages(no more than 1280): ")
        notebook = checker_notebook("Enter a number of pages of each notebook(16, 24 or 32): ")
        if book % notebook != 0:
            raise ValueError
        break
    except ValueError:
        print("A number of book pages have to be a multiple of a number of pages of notebook.")
while True:
    try:
        choice = int(input('If you want to use an additional function, enter a "1". Otherwise enter a "0". Make your choice:'))
        if choice != 1 and choice != 0:
            raise ValueError
        break
    except ValueError:
        print('You have to enter "0" or "1".')
if choice == 1:
    choice = True
else:
    choice = False
decorator(choice, notebooks(book, notebook))
```
